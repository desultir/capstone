```python
from nltk.tokenize.casual import casual_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import re
import pandas as pd
from string import punctuation
import os

input_filepath="../../data/clean/"
filename="filtered_tweets.csv"
```

```python
tweetdf=pd.read_csv(os.path.join(input_filepath, filename),header=0)
tweetdf
```

```python
by_tweeter = tweetdf.groupby(['lga','from_user_name'], as_index=False).agg({'text':lambda x:' '.join(x)})
#by_tweeter['text'] = by_tweeter.apply(lambda x: casual_tokenize(x['text'], strip_handles=True, reduce_len=True, preserve_case=False), axis=1)
by_tweeter
```

```python
def tokenize(text):
    return casual_tokenize(text, strip_handles=True, reduce_len=True, preserve_case=False)

tfidf = TfidfVectorizer(tokenizer=tokenize, stop_words='english')
tfs = tfidf.fit_transform(by_tweeter['text'])
print(tfs.shape[0])

features = np.array(tfidf.get_feature_names())
```

```python
def top_tfidf_feats(row, features, top_n=25):
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.columns = ['feature', 'tfidf']
    return df

#Return top features amongst documents in rows indentified by indices in grp_ids
def top_mean_feats(tfs, features, grp_ids=None, min_tfidf=0.1, top_n=25):
    if grp_ids:
        D = tfs[grp_ids].toarray()
    else:
        D = tfs.toarray()

    D[D < min_tfidf] = 0
    tfidf_means = np.mean(D, axis=0)
    return top_tfidf_feats(tfidf_means, features, top_n)
```
