```python
tweets_aggdf = tweetdf[['lga','id']].loc[~pd.isnull(tweetdf['lat'])].groupby(['lga']).agg('count').reset_index()
tweets_aggdf.columns = ['lga','total_tweets']
tweets_aggdf.sort_values(by='total_tweets',ascending=False)
```

```python
tweeters_aggdf = tweetdf[['lga','from_user_name']].loc[~pd.isnull(tweetdf['lat'])].groupby(['lga']).agg({'from_user_name': lambda x: x.nunique()}).reset_index()
tweeters_aggdf.columns = ['lga','total_tweeters']
tweeters_aggdf.sort_values(by='total_tweeters',ascending=False)
```

```python
lang_ind_df = pd.merge(tweets_aggdf, tweeters_aggdf, on='lga')
lang_ind_df
```

```python
demographics_df = pd.read_csv('../../data/clean/2016_demographics.csv')
demographics_df.rename(columns={'Tot_P_P_G01': 'population'}, inplace=True)
demographics_df[['lga','population']].sort_values(by='population',ascending=False)
```

```python
lang_ind_df = pd.merge(lang_ind_df, demographics_df[['lga','population']],on='lga')
lang_ind_df['tweets_per_capita'] = lang_ind_df['total_tweets'] / lang_ind_df['population']
lang_ind_df['tweeters_per_capita'] = lang_ind_df['total_tweeters'] / lang_ind_df['population']
```
