```python
import numpy as np
import pandas as pd
import os
base = "../../data/raw/"
pd.set_option('display.max_rows', 200)
```

```python
#tweetdf.to_csv("tweets_w_lga.csv")
tweetdf = pd.read_csv('../../data/clean/tweets_w_lga.csv') # shortcut
tweetdf['latlng'] = list(zip(tweetdf.lat, tweetdf.lng))
```

```python
loctweetdf = tweetdf.loc[tweetdf['lat'].notnull(),:]
loctweetdf
```

```python
# Calculate total days covered by the tweet file to determine criteria for ignoring tweets
tdelta = pd.to_datetime(loctweetdf['created_at'].max()) - pd.to_datetime(loctweetdf['created_at'].min())
tdelta = tdelta.days
tdelta
```

```python
aggdf = tweetdf[['lat','lng','text']].loc[~pd.isnull(tweetdf['lat'])].groupby(['lat','lng']).agg('count').reset_index()
aggdf.rename(columns={'text': 'tweet_count'},inplace=True)
aggdf
```

```python
DAILY_TWEET_CUTOFF=30
aggdf['latlng'] = list(zip(aggdf.lat, aggdf.lng))
badgeo_df = aggdf.loc[aggdf['tweet_count'] > DAILY_TWEET_CUTOFF*tdelta]
badgeo_df
#### IF we decide to keep certain posts from dodgy locations, this will be useful...
#posterdf = aggdf.loc[aggdf['tweet_count'] > DAILY_TWEET_CUTOFF*tdelta].merge(tweetdf,how='left',
#                    on=['lat','lng'])[['from_user_name','lat','lng','text']].groupby(
#                    ['from_user_name','lat','lng']).count().reset_index()
#posterdf
#posterdf.loc[posterdf['text']> 20]
```

Visual inspection of the latutude and longitude in google maps indicates these are likely to be "spam" locations.
There may be more locations with the same issue, or 30 maybe be the wrogn cutoff for a larger dataset (TRISMA).

```python
filtered_tweets_df = tweetdf.loc[~tweetdf.latlng.isin(badgeo_df.latlng)]
```

```python
filtered_tweets_df.to_csv("../../data/clean/filtered_tweets.csv",index=False)
```
