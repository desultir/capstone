```python
import numpy as np
import pandas as pd
import os
from timeit import default_timer as timer

base="../../data/clean/"
pd.set_option('display.max_rows', 150)
```

```python
results = pd.read_csv(base+"filtered_nonlog_target.csv").iloc[:,1:]
results['Model'] = 'Target'
results
```

```python
crimes = results.columns[1:-1].tolist()
crimes
```

```python
#We need a baseline as well to compare
temp_df = pd.DataFrame(columns=results.columns)
for n in np.arange(len(results)):
    temp_df = temp_df.append(results.iloc[n,:])
    for crime in crimes:
        temp_df.at[n,crime] = np.mean(results[crime])
temp_df['Model'] = 'Mean Baseline'
results=results.append(temp_df, ignore_index=True)
results
```

```python
# Append each model to results:
models = {'Twitter 17': '17_derived_features_rfr_preds.csv',
           'T17 + TF-IDF': 'ensemble_twitter17_tfidf_rfr_preds.csv',
           'T18 + Opal': 'ensemble_twitter18_opal_rfr_preds.csv',
           'Demographic 12': 'demographics_rfr_preds.csv',
           'All 31': 'ensemble_31_features_rfr_preds.csv',
           'Best 13': 'ensemble_best_13_rfr_preds.csv'}

for (k,v) in models.items():
    temp_df = pd.read_csv(base+v)
    temp_df.columns = results.columns[:-1]
    temp_df['Model'] = k
    results=results.append(temp_df, ignore_index=True)

results
```

```python
results.to_csv("../../data/clean/results.csv",index=False)
```
